# Jucyharo Démo

```python
from jucyharo import *
```

## Fonctions disponibles
**view_total_na**(dataframe) *Afficher total valeurs manquantes*<br>
**view_one_correlation**(dataframe,colname) *Affiche la corrélation heatmap d'une seule variable*<br>
**view_distributions(dataframe)** *Affiche les distributions de toute les colonnes*<br>
**view_correlations(dataframe, y_colname)** *Affiche les corrélation de toute les colonnes pour la même variable y*<br>
**view_one_categorical(df,x_colname,y_colname)** *Affiche la distribution et la corrélation d'une colonne qualitative avec une colonne qualitative.*

```python
dataframe = pd.read_csv('data/train.csv',index_col='Id')
plt.style.use('fivethirtyeight') #Parametre style des graphs
```

### view_total_na(dataframe)
Affiche le total des valeurs manquantes pour chaque colonne d'un dataframe.<br>

```python
view_total_na(dataframe,figsize=(4,4))
```

## view_one_correlation(dataframe,colname)
Applique une matrice de corrélation sur le datafram et affiche un visualisation des corrélations pour une colonne.

```python
view_one_correlation(dataframe,'GrLivArea',figsize=(3,10))
```

## view_distributions(dataframe)
Affiche les distributions selon le type de colonne

```python
cols_bsmt = list(dataframe.columns[dataframe.columns.str.contains('Bsmt')])
view_distributions(dataframe[cols_bsmt])
```

## view_correlations(dataframe, y_colname)
Affiche les corrélations des colonnes du dataframe en x avec une colonne en axe y.

```python
cols_bsmt_price = cols_bsmt + ['SalePrice']
view_correlations(dataframe[cols_bsmt_price],'SalePrice',ncols=4)
```

```python
list_index_outliers = [1299,707]
cols_area = list(dataframe.columns[dataframe.columns.str.contains('Area')])
dataframe_area = dataframe[cols_area+['SalePrice']]

view_correlations(dataframe_area,'SalePrice',outliers=list_index_outliers,ncols=3)
```

# view_one_categorical(df,x_colname,y_colname)
Affiche en même temps la distribution et la corrélation d'une variable catégorique avec une variable quantitative

```python
view_one_categorical(dataframe,'BsmtCond','SalePrice')
```
