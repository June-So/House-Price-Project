# HOUSE PRICE


### À utiliser
**OverallQual** Plus les matériaux utilisé sont de bonnes qualités plus la maison prend de la valeur<br>
**OverallCond** Si la maison est en mauvaise condition elle perd en valeur.<br>
**ExterQual** Plus la qualité des matériaux est pauvre, plus la maison perd en valeur.<br>
**TotalBsmtSF** Plus le sous-sol est grand plus y gagne en valeur.<br>
**1stFlrSF** Plus le premier étage est grand, plus la maison prend de la valeur.<br>
**2ndFlrSF** Plus la surface du deuxieme étage est grande, plus la maison prend de la valeur. Si il n'y a pas de 2eme étage la maison n'en perds pas.<br>
**GrLivArea** Plus la surface habitable au dessus du sol est grande, plus la maison prend de la valeur.<br>
**HeatingQC** Plus la qualité est bonne plus la maison gagne en valeur<br>
**CentralAir** La présence d'air conditionné augmente la valeur de la maison<br>
**Fireplaces** Avoir une ou plusieurs cheminées augmente la valeur de la maison<br>
**FireplaceQu** Si la condition de la cheminée est mauvaise,la maison perd de la valeur<br>
**BsmtFullBath** Le nombre de salle de bain en dessous du sol augmente la valeur<br>
**FullBath** 3 salles de bain au dessus du sol augmente grandement la valeur<br>
**KitchenQual** Plus la qualité de la cuisine est bonne, plus la maison prend de la valeur.<br>
**TotRmsAbvGrd** Plus il y a de pièces, plus la valeur de la maison augmente<br>
**GarageYrBlt** Plus l'année est récente plus la maison peut prendre de la valeur<br>
**GarageFinish** Pas de garage ou un garage non fini peut faire perdre de la valeur<br>
**GarageCars** Plus un garage à de capacité de place plus il peut prendre en valeur.<br>
**GarageQual** Une mauvaise qualité de garage peut faire perdre de la valeur<br>
**GarageCond** un garage en mauvaise condition peut faire perdre de la valeur<br>
**PoolQC** Une piscine en excellente condition fait gonflée la valeur de la maison<br>
**Neighborhood** Certains quartier sont clairement plus cher

**RoofStyle** Un "Hip" donne un peu plus de valeur.<br>

**BedroomAbvGr** : quand il y a 0 chambre au dessus du sol la valeur semble augmentée. Peut-on en conclure que les chambres souterraines apporte de la valeur. /!\ Problème: petite population<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df = pd.read_csv('train.csv')
plt.style.use('fivethirtyeight')
pd.options.display.max_columns = 100

rcParams['figure.figsize'] = 10,6
```

## OverallQual
Plus les matériaux utilisés sont de bonne qualité plus la maison prend de la valeur

```python
df[['OverallQual','SalePrice']].boxplot(by='OverallQual',figsize=(10,6))
```

## OverallCond
Une mauvaise condition de la maison fait perdre de la valeur<br>
Nouvelle variable -> mauvaise condition

```python
df[['OverallCond','SalePrice']].boxplot(by='OverallCond',figsize=(10,6))
```

## YearBuilt

```python
df.plot(kind="scatter",x='YearBuilt', y="SalePrice",figsize=(10,6))
```
